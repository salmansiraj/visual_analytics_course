```python
# mData = table[['Last Update', 'Confirmed', 'Recovered', 'Deaths']]
# dataFrame = pd.DataFrame({
#     'confirmed': table['Confirmed'].head(10).sum(),
#     'recovered': table['Recovered'].head(10).sum(),
#     'deaths': table['Deaths'].head(10).sum()
# }, index = table['Last Update'].unique())
# lines = dataFrame.plot.line()
# -------------- My version above

uniq_dates = list(table['Last Update'].unique())
confirmed = []
recovered = []
deaths = [] 
for date in uniq_dates:
    confirmed.append(table[table['Last Update']== date]['Confirmed'].sum())
    recovered.append(table[table['Last Update']== date]['Recovered'].sum())
    deaths.append(table[table['Last Update']== date]['Deaths'].sum())

    
```

```python
plot = pd.DataFrame()
plot['Last Update'] = uniq_dates
plot['Confirmed'] = confirmed
plot['Recovered'] = recovered
plot['Deaths'] = deaths

plot.head(10)


sns.lineplot(data=plot, x = 'Last Update', y = "Confirmed")
```

```python
line_plot = plot.set_index('Last Update')

sns.lineplot(data=line_plot)
```
