```python
import altair as alt 
import pandas as pd 
```

```python
df = pd.read_csv(filepath_or_buffer="restaurants.csv", sep=',',header=0, index_col=None, lineterminator='\n')
dfboro= df.groupby('BORO').size()
dfboro = dfboro.reset_index(name='count')
alt.Chart(dfboro).mark_bar().encode(
   x='BORO:N',
   y='count:Q'
)
```

```python
alt.Chart(df.head(1000)).mark_bar().encode( 
    x='grades',
    y='# of grades in each',
    color='by GRADE', 
    column = 'by borou'
)
```
